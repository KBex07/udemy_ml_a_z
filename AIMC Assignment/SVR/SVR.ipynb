```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Car_Purchasing_Data.csv', sep=',' , encoding='latin-1')
#there was a UnicodeDecodeError and it was resolved by changing the encoding to latin-1
```

```python
#Observing the dataset
dataset
```

```python
#Plotting graphs between annual salary, cc debt, net worth and car amt
X = dataset.iloc[ : , 5:-1]
y = dataset.iloc[ : , 8]
```

```python
#fit_transform method in FS expects a 2D array => transform y into a 2D array

y = y.values.reshape(-1, 1)    #An attribute error was reported 'Series' object has no attribute 'reshape' => change in pd documentation

print(y)
```

```python
#Splitting into training and testing datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler() #use two diff scaler objects for the feature/dep_var since, the scaler objects notes the mean and median for the given input, which is different for these columns
X_train = sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train,y_train)
```

```python
#Predicting test results
sc_y.inverse_transform(regressor.predict(sc_X.transform(X_test)))
```
