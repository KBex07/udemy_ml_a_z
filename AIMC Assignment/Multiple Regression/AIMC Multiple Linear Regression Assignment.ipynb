```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Import dataset
dataset = pd.read_csv('Train.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

```python
#Training MLR model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X, y)
```

```python
#Import test dataset
dataset = pd.read_csv('Test.csv')
X_test = dataset
```

```python
#Prediction of test set results
y_pred = regressor.predict(X_test)
```

```python
#Print results in a vertical format 
#reshape used to display them vertically
#second argument of concatenate function is 0/1 => vert/horiz concatenation. since 2 vert columns will join to give horiz => 1
np.set_printoptions(precision = 2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)), 1))
```

```python
#Making a single prediction
print(regressor.predict([[1.0152538748373274,2.0762094382543737,-0.26643481961981624,-2.418088204228289,-0.9805027900004764]]))
#values of the features were all input in a double pair of square brackets
#Since "predict" method always expects a 2D array as the format of its inputs
```

```python
#Final Linear Regression equation with coeff values
print(regressor.coef_)
print(regressor.intercept_)
```
